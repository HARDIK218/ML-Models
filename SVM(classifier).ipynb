```python
import numpy as np
```

```python
class svm_classifier():
  def __init__(self, learning_rate, no_of_iteration, lambda_parameter):
    self.learning_rate  = learning_rate
    self.no_of_iteration = no_of_iteration
    self.lambda_parameter = lambda_parameter

  def fit(self, X, Y):
    self.m, self.n = X.shape

    self.w = np.zeros(self.n)
    self.b = 0

    self.X = X
    self.Y = Y

    # implementing gradient descent
    for i in range(self.no_of_iteration):
      self.update_weights()


  def update_weights(self):
    y_label = np.where(self.Y <= 0, -1, 1)
    for index, x_i in enumerate(self.X):
      condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

      if(condition == True):
        dw = 2* self.lambda_parameter * self.w
        db = 0
      else:
        dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
        db = y_label[index]

      self.w = self.w - self.learning_rate * dw
      self.b = self.b - self.learning_rate * db


  def predict(self, X):
    output = np.dot(X, self.w) - self.b
    predicted_labels = np.sign(output)

    y_hat = np.where(predicted_labels <= -1, 0, 1)
    return y_hat

```

```python
classifier = svm_classifier(learning_rate = 0.001, no_of_iteration = 1000, lambda_parameter = 0.01)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data  = pd.read_csv("/content/diabetes_scratch.csv")
```

```python
features = data.drop(columns = 'Outcome', axis = 1)
target = data['Outcome']
```

```python
scaler = StandardScaler()
scaler.fit(features)
features = scaler.fit_transform(features)
```

```python
features
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state=42)
```

```python
#trainig the model
classifier.fit(x_train, y_train)
```

```python
#Accuracy on training data
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
#Accuracy on training data
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(test_data_accuracy)
```
